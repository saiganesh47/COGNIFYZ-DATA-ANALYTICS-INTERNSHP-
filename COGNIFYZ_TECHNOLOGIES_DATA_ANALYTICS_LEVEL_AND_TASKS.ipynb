```python
import pandas as pd
import nltk
nltk.download('punkt')
nltk.download('stopwords')
nltk.download('vader_lexicon')
from nltk.sentiment.vader import SentimentIntensityAnalyzer
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
from collections import Counter
```

```python
data_set = pd.read_csv('/content/Dataset .csv')
data_set
```

```python
data_set.head(5)
```

```python
data_set.tail(5)
```

```python
rating_texts = data_set['Rating text']
print(rating_texts)
```

```python
sia=SentimentIntensityAnalyzer()
stopwords=set(stopwords.words('english'))
positive = []
negative = []
```

```python
for rating_text in rating_texts:
    tokens = word_tokenize(rating_text.lower())
    tokens = [token for token in tokens if token.isalpha() and token not in stopwords]
    sentiment_score = sia.polarity_scores(rating_text)['compound']
    if sentiment_score >= 0.2:
      positive.extend(tokens)
    elif sentiment_score<=0.2:
      negative.extend(tokens)


```

```python
positive_counts = Counter(positive)
negative_counts = Counter(negative)
```

```python
num_top_keywords = 20
print('\nTop positive:')
for keyword, count in positive_counts.most_common(num_top_keywords):
    print(f"{keyword}:{count} times")
```

```python
num_top_keywords = 20
print('\nTop negative:')
for keyword, count in negative_counts.most_common(num_top_keywords):
    print(f"{keyword}:{count} times")
```

```python
average_review = data_set[['Rating text', 'Aggregate rating']].copy()
average_review['Review Length'] = average_review['Rating text'].apply(lambda x: len(x))
average_review_length = average_review.groupby('Aggregate rating')['Review Length'].mean()

```

```python
import matplotlib.pyplot as plt
import seaborn as sns
plt.figure(figsize=(10,10))
average_review_length.plot(kind='bar',color='green')
plt.title('Average Review Length For Each Rating Category')
plt.xlabel('Aggregate Rating')
plt.ylabel('Average review length character')
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
data_set = pd.read_csv('/content/Dataset .csv')
data_set.head(2)
```

```python
data_set.tail(3)
```

```python
data_set_analysis=data_set[['Price range','Has Online delivery','Has Table booking']].copy()
data_set_analysis['Has Online delivery']=data_set_analysis['Has Online delivery'].map({'Yes':True,'No':False})
data_set_analysis['Has table booking']=data_set_analysis['Has Table booking'].map({'Yes':True,'No':False})
summary_table=pd.pivot_table(data_set_analysis,index='Price range',values=['Has Online delivery','Has table booking'],aggfunc=sum)
print(summary_table)
```

```python
plt.figure(figsize=(10,8))
summary_table.plot(kind='bar',stacked=True,colormap='cividis')
plt.title('Relationship between price Range and Availibility')
plt.xlabel('Price range')
plt.ylabel('Count')
plt.legend(title='Feature',loc='upper left')
plt.show()
```

```python
delivery_proportions = data_set.groupby('Price range')['Has Online delivery'].value_counts(normalize=True).unstack().fillna(0)
booking_proportions = data_set.groupby('Price range')['Has Table booking'].value_counts(normalize=True).unstack().fillna(0)
print("Online Delivery Proportions:\n", delivery_proportions)
print("\nTable Booking Proportions:\n", booking_proportions)
```

```python
plt.figure(figsize=(10,6))
plt.subplot(1,2,1)
sns.countplot(x='Price range' , hue='Has Online delivery' , data=data_set_analysis,palette='copper')
plt.title('Online Delivery Avalibility by Price Range')
plt.tight_layout()
plt.subplot(1,2,2)
sns.countplot(x='Price range', hue='Has Table booking', data=data_set_analysis,palette='gist_earth')
plt.title('Table Booking Availibility by Price range')
plt.tight_layout()
plt.show()
```
